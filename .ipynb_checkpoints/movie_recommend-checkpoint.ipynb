```python
# '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'

# {'Action','Adventure','Fantasy','Sci-fi'}
import ast # importing ast library
```

```python
def convert(obj):
    l = []
    for i in ast.literal_eval(obj): # ast.literal_eval converts string of lists into list
        l.append(i['name'])
    return l
```

```python
movies['genres']= movies['genres'].apply(convert)
```

```python
movies.head(2)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies['cast'][0]
```

```python
def convert2(obj):
    l = []
    counter=0
    for i in ast.literal_eval(obj): # ast.literal_eval converts string of lists into list
        if counter != 3:
          l.append(i['name'])
          counter+=1
    return l
```

```python
movies['cast']=movies['cast'].apply(convert2)
```

```python
movies.head(2)
```

```python
def fetch_director(obj):
    l = []
    for i in ast.literal_eval(obj): # ast.literal_eval converts string of lists into list
        if i['job'] == 'Director':
          l.append(i['name'])
          break
    return l
```

```python
movies['crew']= movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split()) #Python provides the built-in split() function to split strings into separated pieces. The split() function separates a string into multiple strings, arranges them in a list, and returns the list.
```

# we ll use bag of words technique to convert out text to vectors 

# Bag of Words is a method to extract features from text documents. These features can be used for training machine learning algorithms. It creates a vocabulary of all the unique words occurring in all the documents in the training set. For example, if you have 3 documents-

D1 - “I am feeling very happy today”

D2 - “I am not well today”

D3 - “I wish I could go to play”

First, it creates a vocabulary using unique words from all the documents -

Unique list of words -

I am feeling very happy today not well wish could go to play

Then, for each word the frequency of the word in the corresponding document is inserted



The above table depicts the training features containing term frequencies of each word in each document. This is called bag-of-words approach since the number of occurrence and not sequence or order of words matters in this approach.

```python
# nltk is used to stemm words like loved loves to love so that their no reccurrance , its explanation with example is below
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
new_df['tags']= new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words = 'english') # stop words = 'english' means ignoring words like is am are to from etc
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

# Link to explain cosine similarity -> https://medium.datadriveninvestor.com/cosine-similarity-cosine-distance-6571387f9bf8


```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True #desc order , key=lambda x:x[1] # we need to sort acc to similarity not index )[1:6]  # Enumerate() method adds a counter to an iterable and returns it in a form of enumerating object. This enumerated object can then be used directly for loops or converted into a list of tuples using the list() method.
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.loc[i[0]].title) # i[0] is the index from the recommended movies and we fetch the title of all the recommended movies through their indexes  
```

```python
recommend('Batman Begins') # lets check 
```
